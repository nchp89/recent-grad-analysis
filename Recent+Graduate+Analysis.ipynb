```python
#drop rows that are missing data
recent_grads = recent_grads.dropna()
```

```python
#make some scatter plots to explore the data
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs Median', figsize=(5,10))
```

## The median salary seems to actually decline as the number of full-time workers increase. This suggests that overly-saturated areas of study lead to lower-paying employment. Most likely due to competition and supply/demand.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs Unemployment_rate', figsize=(5,10))
```

## There is no correlation between unemployment and the share of women who studied a given major.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs Median', figsize=(5,10))
```

## As the number of men in a given field of study increase, the median salary for that field of study approaches ~40,000

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs Median', figsize=(5,10))
```

## As the number of women in a given field of study increase, that field of study's median salary approaches ~35,000 dollars. 

## Conclusion: Men tend to choose majors that lead to higher paying jobs. 

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
plt.title('Sample Size')
```

```python
recent_grads['Median'].hist(bins=40, range=(0,80000))
plt.title('Median')
```

## The most common median salary is 34,000 to 36,000

```python
recent_grads['Employed'].hist(bins=25)
plt.title('Employed')
```

```python
recent_grads['Full_time'].hist(bins=25)
plt.title('Full Time')
```

```python
recent_grads['ShareWomen'].hist(bins=25)
plt.title('Share of Women')
```

## From this histogram, we can see that most majors are predominantly female. This is a bit surprising, but if we do some math we can see that women outnumber men in the data by a ratio of 1.35

```python
total_men = recent_grads['Men'].sum()
total_women = recent_grads['Women'].sum()
ratio = total_women/total_men
print('{}:Men\n{}:Women'.format(total_men, total_women))
print('Ratio of women to men: {}'.format(ratio))
```

## From the above two figures, we can see that the lowest-paying majors have higher shares of women

```python
recent_grads.plot.bar(x=['Major_category'], y=['Men','Women'], figsize=(30,5), title=('Quantity of Genders in Each Major'))
```

## Recall that Major_category is sorted by median salary. Notice that as the Major_category progresses (the x-axis) the amount of women students seem to increase. This shows that women choose lower-paying majors more often than men.
